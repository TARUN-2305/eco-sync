### Phase B: Reinforcement Learning (PPO)

**Methodology:**
With the `BengaluruBusEnv` established, we deploy Proximal Policy Optimization (PPO) using `stable-baselines3`. PPO is chosen for its sample efficiency and stability in continuous/discrete action spaces. The agent plays thousands of simulated episodes, iteratively learning when to dispatch, hold, or skip stops to maximize the long-term economic reward (thereby inherently mitigating bus bunching).

```
import os
import sys
sys.path.append("..") # To allow importing from parent directory if needed

import gymnasium as gym
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv

# Import custom environment
from test_env import BengaluruBusEnv

os.makedirs("../models", exist_ok=True)
env = BengaluruBusEnv()
vec_env = DummyVecEnv([lambda: env])

model = PPO("MlpPolicy", vec_env, verbose=1)
model.learn(total_timesteps=10000)

model.save("../models/eco_sync_ppo")
print("Model saved successfully to models/eco_sync_ppo.zip")
```
